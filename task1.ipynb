```python
import pandas as pd

data = pd.read_csv("business_performance_dataset.csv")
data.head()
```

```python
data.describe()
```

```python
data['Region'].value_counts()
```

```python
data.groupby('Product_Category')['Revenue_INR'].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(data['Date'], data['Revenue_INR'])
plt.xticks(rotation=45)
plt.title("Revenue Trend Over Time")
plt.show()
```

```python
import matplotlib.pyplot as plt

region_revenue = data.groupby("Region")["Revenue_INR"].sum()

plt.figure()
region_revenue.plot(kind="bar")
plt.title("Total Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue (INR)")
plt.show()
```

```python
data.groupby("Product_Category")["Units_Sold"].sum().plot(kind="bar")
plt.title("Units Sold by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Units Sold")
plt.show()
```
